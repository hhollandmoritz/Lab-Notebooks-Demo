## Using Jupyter Notebooks Remotely ##

Let's say your data is on a remote server (i.e. Microbe) and you want to interact with it through a jupyter notebook. To set this up, follow the steps below:

*This tutorial was adapted from this website: https://amber-md.github.io/pytraj/latest/tutorials/remote_jupyter_notebook *

**Pre-requisites** - make sure jupyter notebook is installed on your own computer and on the server you are trying to connect with.

** Part 1: On the remote server **

1. Open a terminal window and log onto the remote server

    <code>ssh username@microbe.colorado.edu</code>

2. ```cd``` into the directory with your notebook (or the directory where you want to create a new notebook)
3. type ```jupyter notebook --no-browser --port=1234```

    **Explanation:**

    <code>--no-browser</code>

    This command tells jupyter notebook not to try to launch a browser window, which is good since the default "browser" in a terminal window, is pretty horrible.
    
    <code>--port=1234</code>

    This command tells jupyter notebook that instead of opening a browser it should send the notebook to a special outlet called a port. Ports are specified by numbers, the numbers can be almost anything. It is probably a good idea to make sure that multiple users do not use the same port numbers since conflicts can arrise if two people request the same port at the same time.

** Part 2: On your local computer **

4. Open a second terminal window. On your local computer type
<code>ssh -N -f -L localhost:8888:localhost:1234 username@your_remote_host_name</code>

 **Explanation:**

    <code>-N</code>

    This command works with the ```-f``` command to help the ssh command run in the background.
    
    <code>-f</code>

    This command tells the terminal to put ssh command in the background just before it runs. This is mainly so that after you run the command, you can still use the terminal that you ran it in.
    
    <code>-L</code>

    This command specifies the port-to-port connections. It is followed by a ```localhost:8888:localhost:1234``` you can read this as instructions to the computer. "Connect port 8888 on the current host to port 1234 on the remote host."
    
5. Type in your server password when prompted.

6. Open a web browser (chrome, firefox, etc.) and type localhost:8888 in the address line.

** Don't forget **

You're running the notebook on the server so if you choose to run R in your notebook, you'll need to install all of the R packages on the server before you do that.

## Making Juypter Notebooks Pretty ##

To make a juypter notebook "pretty" (i.e. formatted the way you want) you need to learn markdown. 

Markdown is a simple set of text-only structures that can be interpreted as formatting. It's a simple way to convert a text file into a formated document. 

An in-depth guide can be found here: https://daringfireball.net/projects/markdown/syntax#p

Below are some basic examples:

### Markdown Basics ###

# Header 1
## Header 2
### Header 3 
etc..

Making text *italic* or **Bold** or ***Bold and italic!***

Bullet points work like this:

* first point
* second point
* third point
    * sub point of three
* Back to 4th point

*Importantly!: in bulletted lists, make sure to put a space between the bullet symbol and the text, otherwise the list will not look right and also be sure to put an empty line before and after your list to make sure the interpreter knows when to stop the list.*

Dashes also work the same way:

- Point number 1
- 2
- 3
    - sub 3
    * bullet character change won't make a difference here
- 4

Numbered lists work like this:

1. Number 1
2. Number 2
3. Number 3
    1. Sub list of 3
4. Number 4

Below is a way to make a horizontal line:
__________________________________________

### Blockquotes ###

>
